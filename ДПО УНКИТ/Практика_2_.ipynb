## Задание 2

```python
list1 = list(map(int, input("Введите 1ый список целых чисел через пробел!: ").split()))
list2 = list(map(int, input("Введите 2ой список целых чисел через пробел!: ").split()))

print("1ый список:", list1)
print("2ой список:", list2)

# список для взятия с двух других.
result_list = []


result_list.extend(list1[i] for i in range(0, len(list1), 2))

result_list.extend(list2[i] for i in range(1, len(list2), 2))

print("Итоговый список:", result_list)
```

## Задание 3

```python
input_list = input("Введите элементы списка через пробел: ").split()

# преобразуем элементы
processed_list = []
for item in input_list:
    try:
        # целое число
        processed_list.append(int(item))
    except ValueError:
        try:
            # число с плавающей точкой
            processed_list.append(float(item))
        except ValueError:
            # не удалось=оставляем как строку
            processed_list.append(item)

# список
print("Исходный список:", processed_list)

# удаление дубликатов
unique_list = []
seen = set()  # множество для отслеживания уже добавленных элементов
for item in processed_list:
    # приводим элемент к строке для сравнения
    str_item = str(item)
    if str_item not in seen:
        seen.add(str_item)
        unique_list.append(item)

print("Список без дубликатов:", unique_list)
```

## Задание 4

```python
input_dict = {}
n = int(input("Введите количество элементов словаря: "))

for _ in range(n):
    key = input("Введите ключ (строку): ")
    value = input("Введите значение (целое число или число с плавающей точкой): ")

    # знач. в int или float
    try:
        value = int(value)
    except ValueError:
        try:
            value = float(value)
        except ValueError:
            print("Ошибка: значение должно быть целым числом или числом с плавающей точкой.")
            exit()
    input_dict[key] = value

print("Исходный словарь:", input_dict)

# создание списка кортежей
value_to_keys = {}  # словарь для группировки ключей по значениям
for key, value in input_dict.items():
    if value not in value_to_keys:
        value_to_keys[value] = []
    value_to_keys[value].append(key)

# создание списка кортежей
result_list = [(value, keys) for value, keys in value_to_keys.items()]

print("Список кортежей (значение, список ключей):", result_list)
```

## Задание 5

```python
def input_dict():
    """Функция для ввода словаря с клавиатуры."""
    n = int(input("Введите количество элементов словаря: "))
    input_dict = {}
    for _ in range(n):
        key = input("Введите ключ (строку): ")
        value = input("Введите значение (целое число или число с плавающей точкой): ")

        # знач. в int или float
        try:
            value = int(value)
        except ValueError:
            try:
                value = float(value)
            except ValueError:
                print("Ошибка: значение должно быть целым числом или числом с плавающей точкой.")
                exit()
        input_dict[key] = value
    return input_dict

print("Введите первый словарь:")
dict1 = input_dict()

print("Введите второй словарь:")
dict2 = input_dict()

print("Первый словарь:", dict1)
print("Второй словарь:", dict2)

# пересечение множеств значений
values1 = set(dict1.values())
values2 = set(dict2.values())
intersection_values = values1.intersection(values2)

# словарь пересечений
result_dict = {}
for key, value in dict1.items():
    if value in intersection_values:
        result_dict[key] = value
for key, value in dict2.items():
    if value in intersection_values:
        result_dict[key] = value

print("Пересечение значений:", intersection_values)
print("Новый словарь с общими значениями:", result_dict)
```
